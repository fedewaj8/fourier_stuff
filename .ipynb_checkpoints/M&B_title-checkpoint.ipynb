```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.fft import fft, ifft, fftfreq
from playsound import playsound
import math
```

```python
# playsound('titlescreen.wav')
```

```python
from scipy.io import wavfile
samplerate, data = wavfile.read('titlescreen.wav');

print(f"number of channels = {data.shape[1]}")
length = data.shape[0] / samplerate
print(f"length = {length}s")
```

```python
length,len(time),samplerate
```

```python
time = np.linspace(0., length, data.shape[0])

fig = plt.figure(figsize=(8,6))
plt.plot(time, data[:, 0], label="Left channel")
plt.plot(time, data[:, 1], label="Right channel",alpha=0.7)
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()
```

```python
T0 = length
N = len(time)
dt = T0/N
```

```python
#left channel
L = data[:,0]
R = data[:,1]
```

```python
plt.plot(time,L)
plt.plot(time,R)
```

```python
xf = fftfreq(N,dt)
Lyf = (1/N)*fft(L)
Ryf = (1/N)*fft(R)
```

```python
plt.plot(xf[0:N//2],abs(Lyf[0:N//2]),label='Left')
plt.plot(xf[0:N//2],abs(Ryf[0:N//2]),label='Right')
plt.xlabel('frequency')
plt.ylabel('amplitude')
# plt.xlim(435,455)
plt.grid()
plt.legend()
```

```python
plt.plot(xf[0:N//2],abs(Lyf[0:N//2]),label='Left')
plt.plot(xf[0:N//2],abs(Ryf[0:N//2]),label='Right')
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.xlim(0,200)
plt.grid()
plt.legend()
```

```python
plt.plot(xf[0:N//2],abs(Lyf[0:N//2]),label='Left')
plt.plot(xf[0:N//2],abs(Ryf[0:N//2]),label='Right')
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.xlim(435,455)
plt.grid()
plt.legend()
```

```python
freq = xf
LVtrimf = Lyf

LVtrimf[ (freq < -455) | (freq > 455)] = 0
LVtrimf[np.where((freq > -435) & (freq < 435))] = 0
LVtrim = ifft(N*LVtrimf)

plt.figure(figsize=(1.618*5,5))
plt.plot(time,L,label='original Left')
plt.plot(time,np.real(LVtrim)+20000,label='trimmed Left')
plt.xlabel('time (s)')

plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(1.618*5,5))
plt.plot(time,L,label='original Left')
plt.plot(time,(np.real(LVtrim)).astype(int)+20000,label='trimmed Left')
plt.xlabel('time (s)')

plt.grid()
plt.legend()
```

```python
np.real(LVtrim)
```

```python
wavfile.write('left445.wav',samplerate,(np.real(LVtrim)).astype(np.int16))
```
