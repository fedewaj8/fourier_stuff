```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.fft import fft, ifft, fftfreq
from playsound import playsound
import math
```

```python
from scipy.io import wavfile
samplerate, data = wavfile.read('titlescreen.wav');

print(f"number of channels = {data.shape[1]}")
length = data.shape[0] / samplerate
print(f"length = {length}s")
```

```python
time = np.linspace(0., length, data.shape[0])

fig = plt.figure(figsize=(8,6))
plt.plot(time, data[:, 0], label="Left channel")
plt.plot(time, data[:, 1], label="Right channel",alpha=0.7)
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()
```

```python
T0 = length
N = len(time)
dt = T0/N
```

```python
L = data[:,0] #left channel
R = data[:,1] #right channel
```

```python
xf = fftfreq(N,dt)
Lyf = (1/N)*fft(L)
Ryf = (1/N)*fft(R)
```

```python
plt.plot(xf[0:N//2],abs(Lyf[0:N//2]),label='Left')
plt.plot(xf[0:N//2],abs(Ryf[0:N//2]),label='Right')
plt.xlabel('frequency')
plt.ylabel('amplitude')
# plt.xlim(435,455)
plt.grid()
plt.legend()
```

# Converting back to .wav file

```python
freq = xf
LVtrimf = Lyf

LVtrim = ifft(N*LVtrimf)

plt.figure(figsize=(1.618*5,5))
plt.plot(time,L,label='original Left')
plt.plot(time,np.real(LVtrim)+20000,label='trimmed Left')
plt.xlabel('time (s)')

plt.grid()
plt.legend()
```

```python
wavfile.write('original.wav',samplerate,(np.real(LVtrim)).astype(np.int16))
```

# Isolating frequencies within the range of a trumpet (165Hz < $f$ < 932Hz)

```python
xf = fftfreq(N,dt)
Lyf = (1/N)*fft(L)
Ryf = (1/N)*fft(R)

plt.plot(xf[0:N//2],abs(Lyf[0:N//2]),label='Left')
plt.plot(xf[0:N//2],abs(Ryf[0:N//2]),label='Right')
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.xlim(165,932)
plt.grid()
plt.legend()
```

```python
freq = xf
LVtrimf = Lyf

LVtrimf[ (freq < -932) | (freq > 932)] = 0
LVtrimf[np.where((freq > -165) & (freq < 165))] = 0
LVtrim = ifft(N*LVtrimf)

plt.figure(figsize=(1.618*5,5))
plt.plot(time,L,label='original Left')
plt.plot(time,np.real(LVtrim)+20000,label='trimmed Left')
plt.xlabel('time (s)')

plt.grid()
plt.legend()
```

```python
wavfile.write('trumpet.wav',samplerate,(np.real(LVtrim)).astype(np.int16))
```
